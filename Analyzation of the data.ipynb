```python
# Importing libraries
import h5py
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Analyzing the dataset

```python
# Opening the HDF5 file
f = h5py.File('web-scraping','r')
```

```python
# Retrieving population data
total_number_of_popualtion_for_each_day = []
days = list(range(26,32))
for day in days:
    world_population = f[f'World population/World population 2023-05-{day}']
    total_number_of_popualtion_for_each_day.append(sum(world_population['country population']))
```

```python
# Analyzing population data
fig , ax = plt.subplots(figsize=(10,5))
sns.barplot(days,total_number_of_popualtion_for_each_day)
ax.set_yscale('log')
ax.set_title('Total number of population in all 20 countries for the date of May 2023')
ax.set_xlabel('Day')
ax.set_ylabel('Number of population')
plt.show()
```

```python
# Retrieving investment data
records = []
for day in days:
    records.append(f[f'Investment/Investment 2023-05-{day}'][:])
names = []
prices = []
for record in range(0,6):
    for value in records[record]:
        names.append(value[0])
        prices.append(value[1])
investment_df = pd.DataFrame({'company names':names,'company prices':prices})
```

```python
# Analyzing investment data
tesla = investment_df[investment_df['company names'] == b'Tesla, Inc.']
tesla_prices = tesla['company prices']
palantir_tech = investment_df[investment_df['company names'] ==  b'Palantir Technologies Inc.']
palantir_tech_prices = palantir_tech['company prices']
fig , (ax1,ax2) = plt.subplots(1,2,figsize=(15,7))
ax1.plot(days,tesla_prices,linestyle='-',marker='o')
ax1.set_title('Change of prices for Tesla company in May 2023')
ax1.set_xlabel('Day')
ax1.set_ylabel('Price in Billions USD')
ax2.plot(days,palantir_tech_prices,linestyle='-',marker='o')
ax2.set_title('Change of prices for Palantir technologies company in May 2023')
ax2.set_xlabel('Day')
ax2.set_ylabel('Price in Billions USD')
plt.show()
```

```python
# Retrieving crypto currency data
records = []
for day in days:
    crypto = f[f'Crypto currency/Crypto currency 2023-05-{day}']
    records.append(crypto[:])
bitcoin = []
ethereum = []
thether = []
BNB = []
for x in records:
    bitcoin.append(x[0])
    ethereum.append(x[1])
    thether.append(x[2])
    BNB.append(x[3])
bitcoin_marketcap = []
ethereum_marketcap = []
thether_marketcap = []
BNB_marketcap = []
for value in bitcoin:
    bitcoin_marketcap.append(value[1])
for value in ethereum:
    ethereum_marketcap.append(value[1])
for value in thether:
    thether_marketcap.append(value[1])
for value in BNB:
    BNB_marketcap.append(value[1])
```

```python
# Analyzing crypto currency data
fig , ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(18,10))
ax1.bar(days,bitcoin_marketcap)
ax2.bar(days,ethereum_marketcap)
ax3.bar(days,thether_marketcap)
ax4.bar(days,BNB_marketcap)
ax1.set_title('Trends in Bicoin marketcaps for May 2023')
ax2.set_title('Trends in Ethereum marketcaps for May 2023')
ax3.set_title('Trends in Thether marketcaps for May 2023')
ax4.set_title('Trends in BNB marketcaps for May 2023')
ax1.set_yscale('log')
ax2.set_yscale('log')
ax3.set_yscale('log')
ax4.set_yscale('log')
ax1.set_ylabel('Marketcap prices in USD')
ax2.set_ylabel('Marketcap prices in USD')
ax3.set_ylabel('Marketcap prices in USD')
ax4.set_ylabel('Marketcap prices in USD')
ax1.set_xlabel('Days')
ax2.set_xlabel('Days')
ax3.set_xlabel('Days')
ax4.set_xlabel('Days')
```
